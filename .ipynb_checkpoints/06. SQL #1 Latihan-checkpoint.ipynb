```python
#settingan awal, hanya dilakukan sekali saja

import os
import pymysql #koneksi ke mysql
import pandas as pd #tampung koneksi ke panda

host = '127.0.0.1'
port = "3306" #port default mysql
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')
#running sql query
def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

### 1. Prepare a list of offices sorted by country, state, city.

```python
run("select * from offices order by country, state, city")
```

### 2. How many employees are there in the company?

```python
run("select count(*) as numberOfEmployees from employees")
```

### 3. What is the total of payments received?

```python
run("select sum(amount) as totalOfPayments from payments")
```

### 4. List the product lines that contain 'Cars'.

```python
run("select * from productlines where productLine LIKE '%Cars%'")
```

### 5. Report total payments for October 28, 2004.

```python
run("""
   select sum(amount) as totalPayment28Oct2004 from payments where YEAR(paymentDate)='2004' 
   AND MONTH(paymentDate)='10' AND DAY(paymentDate)='28'
""")
 
```

### 6. Report those payments greater than \$100,000.

```python
run("""
    select * from payments where amount > 100000
""")
```


### 7. List the products in each product line.

```python
tProductLines = run(""" select * from productLInes""")
```

```python
tProductLines.productLine[1]
lenProductLines = len(tProductLines)
```

```python
def product():
    for i in range(0,lenProductLines):
        xProductLine = tProductLines.productLine[i]
        print(xProductLine)
        query = "select * from products where productLine='"+xProductLine+"'"
        x = run(query)
        print(x,"\n")
        
product()
```


### 8. How many products in each product line?

```python
run("""
select productLine, count(*) as count from products group by productLine
""")
```



### 9. What is the minimum payment received?

### 11. What is the average percentage markup of the MSRP on buyPrice?

```python
#run("""select * from products""")
run(""" select avg((msrp-buyPrice)/buyPrice)*100 as average from products""")
```

### 12. How many distinct products does ClassicModels sell?

```python
run("""select count(DISTINCT productCode) as count from products""")
```

### 13. Report the name and city of customers who don't have sales representatives?

```python
run("""
select customerName, city 
from customers where length(salesRepEmployeeNumber)>0
""")
```


### 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
#run("select * from employees")
run("""
 select CONCAT(firstName, ' ', lastName) as fullName, jobTitle from Employees
 where jobTitle LIKE '%VP%'or jobTitle like '%Manager%'
 """)
```



### 15. Which orders have a value greater than $5,000?

```python
run("""
select orderNumber, productCode, (quantityOrdered * priceEach) as value, orderLineNumber
from orderDetails 
having value > 5000
""")
```
