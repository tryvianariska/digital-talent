```python
import pandas as pd
```

```python
data_wisatawan_mancanegara = pd.read_csv('wisatawanmancanegarayangdatangkedkijakartamenurutpintumasukdanbulan2012.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_wisatawan_mancanegara.groupby(['tahun', 'pintu_masuk']).mean()
```

```python
data_wisatawan_mancanegara.groupby(['tahun', 'bulan']).mean()
```

```python
ax = data_wisatawan_mancanegara.groupby(['tahun','bulan']).mean()['jumlah']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_wisatawan_mancanegara.groupby('bulan').mean()['jumlah']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Bulan", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data_wisatawan_mancanegara.groupby('pintu_masuk').mean()['jumlah']
ax = ax.plot(kind='bar', title ="Rata-Rata Berdasarkan Pintu Masuk", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
