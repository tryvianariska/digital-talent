```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## One to many relationship

```python
# 1 Report the account representative for each customer.
run("""
select customerNumber, customerName, salesRepEmployeeNumber, concat(firstName, ' ', lastName) as accountRepresentative
from customers C
inner join employees E on C.salesRepEmployeeNumber = E.employeeNumber""")
```

```python
#2.Report total payments for Atelier graphique.
#run("""select * from customers where customerName like '%Atelier%'""")

run("""
select customerNumber, customerName, sum(amount) as total
from customers
inner join payments using(customerNumber)
where customerName = 'Atelier graphique'
group by customerNumber, customerName
""")
```

```python
#3.Report the total payments by date

run("""
select YEAR(paymentDate) as Tahun,
       MONTH(paymentDate) as Bulan,
       DAY(paymentDate) as Hari,
       sum(amount) as Total
from payments
group by tahun, bulan, hari
""")
```

```python
# 4. Report the products that have not been sold.
run("""
select P.productCode, P.productName,O.productCode
from products P
left join orderdetails O using(productCode)
where O.productCode is null
""")
```

```python
# 5.List the amount paid by each customer.
run("""
select customerNumber, customerName, sum(amount) as amount
from customers
left join payments using(customerNumber)
group by customerNumber,customerName
""")
```

```python
# 6. How many orders have been placed by Herkku Gifts?
run("""
select C.customerNumber, C.customerName, sum( OD.quantityOrdered) as totalOrder
from customers C
inner join orders O using(customerNumber) 
inner join orderdetails OD using(orderNumber)

where customerName like '%Herkku%'
group by C.customerNumber, C.customerName
""")
```

```python
# 7. Who are the employees in Boston?
run("""
select city, concat(firstName, ' ', lastName) as employeeName
from offices O
inner join employees using (officeCode)
where city = 'Boston'
""")
```

```python
# 8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.

run("""
select customers.customerNumber, customerName, sum(amount) as totalPayments
from payments
inner join customers using (customerNumber)
group by customers.customerNumber, customerName
having totalPayments > 100000
order by totalPayments desc
""")
```

```python
#9. List the value of 'On Hold' orders.

run("""
select O.orderNumber, sum(quantityOrdered * priceEach) as value
from orders O
inner join orderdetails OD using(orderNumber)
where status ='On Hold'
group by O.orderNumber
""")
```

```python
# 10. Report the number of orders 'On Hold' for each customer.
run("""
select customerName, count(orderNumber) as numberOfOrder
from orders
inner join customers using (customerNumber)
where status = 'On Hold'
group by customerName
""")
```

## Many to many relationship

```python
#1. List products sold by order date  ?productCode, productName,
run("""
select orderDate , sum(quantityOrdered) totalSold
from orders O
inner join orderdetails OD using (orderNumber)
inner join products P using (productCode)
group by orderDate
""")
```

```python
#2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.
run("""
select orderDate
from orders 
inner join orderdetails using (orderNumber)
inner join products using (productCode)
where productName = '1940 Ford Pickup Truck'
order by orderDate desc
""")
```

```python
#3. List the names of customers and their corresponding order number where a particular order from 
#that customer has a value greater than $25,000?

run("""
select customerName, O.orderNumber, sum(quantityOrdered * priceEach) amount
from customers
inner join orders O using (customerNumber)
inner join orderDetails using (orderNumber)
group by customerName, orderNumber
having amount > 25000
order by customerName
""")
```

```python
# 4. Are there any products that appear on all orders?

run("""
select productCode, count(orderNumber) as appear
from orderDetails
group by productCode
having appear = (select count(1) from orders)
""")

#tidak ada
```

```python
#5. List the names of products sold at less than 80% of the MSRP.
run("""
select orderNumber, P.productCode, productName, MSRP,priceEach
from products P
inner join orderdetails OD using (productCode)
where priceEach < (0.8 * MSRP)

""")
```

```python
#6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)
#6.1 asumsi products di list sesuai ordernumbernya, dan products dg ordernumber bersesuaian dengan ratenya yang >= 2 yang dilist
run("""
select productName, orderNumber,buyPrice, priceEach, (priceEach/buyPrice) as rate
from products 
inner join orderDetails using(productCode)
having rate >= 2
""")
```

```python
#6.2 asumsi jika product pernah dijual dengan rate >= 2, product masuk list. rate = (priceEach/buyPrice) 
run("""
select productName
from products 
inner join orderDetails using(productCode)
where (priceEach/buyPrice) >= 2
group by productName


""")
```

```python
#7. List the products ordered on a Monday. 
# weekday = 0 => monday
run("""
select distinct productName, weekday(orderDate) day
from orders o
inner join orderdetails od using (orderNumber)
inner join products p using (productCode)
having day = 0
""")
```

```python
#8. What is the quantity on hand for products listed on 'On Hold' orders?
run("""
select productName, quantityInStock
from orders o 
inner join orderdetails od using (orderNumber)
inner join products p using (productCode)
where status = 'On Hold'
""")
```
